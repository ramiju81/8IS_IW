```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np
import cv2

def create_banner(output_path="banner.png"):
    # Dimensiones del banner
    width, height = 1200, 600

    # Crear un fondo basado en la imagen de referencia
    background = np.zeros((height, width, 3), dtype=np.uint8)
    for y in range(height):
        color_ratio = y / height
        sky_color = np.array([10, 10, 50])  # Azul oscuro espacial
        earth_color = np.array([50, 25, 0])  # Marrón tierra
        mixed_color = (1 - color_ratio) * sky_color + color_ratio * earth_color
        background[y, :] = mixed_color

    # Agregar la luna y la tierra en la parte inferior
    moon_radius = 200
    cv2.circle(background, (width//2, height+moon_radius//2), moon_radius, (200, 200, 200), -1)

    # Convertir fondo a imagen PIL
    background = Image.fromarray(cv2.cvtColor(background, cv2.COLOR_BGR2RGB))

    # Cargar la imagen del robot (se espera que el usuario la proporcione como "image.png")
    try:
        robot = Image.open("/mnt/data/image.png").resize((250, 250))
        background.paste(robot, (100, 250), robot if robot.mode == 'RGBA' else None)
    except Exception as e:
        print(f"Error al cargar la imagen del robot: {e}")

    # Dibujar una burbuja de pensamiento correctamente orientada
    draw = ImageDraw.Draw(background)
    bubble_x, bubble_y = 400, 150
    draw.ellipse([bubble_x, bubble_y, bubble_x+250, bubble_y+180], fill=(255, 255, 255), outline=(0, 0, 0))
    draw.ellipse([bubble_x-50, bubble_y+100, bubble_x-20, bubble_y+130], fill=(255, 255, 255), outline=(0, 0, 0))
    draw.ellipse([bubble_x-80, bubble_y+130, bubble_x-50, bubble_y+160], fill=(255, 255, 255), outline=(0, 0, 0))

    # Insertar texto dentro de la burbuja
    font = ImageFont.load_default()
    draw.text((bubble_x+40, bubble_y+40), "Innovación\nAutomatización\nCreatividad\nOptimización", fill=(0, 0, 0), font=font)

    # Agregar el texto principal al lado derecho
    text_x, text_y = 650, 100
    paragraph = (
        "La innovación no es un destino, es un viaje constante entre la realidad y el futuro.\n\n"
        "Cada línea de código es una chispa de creatividad, cada algoritmo un motor de cambio.\n"
        "Entre datos y automatización, el ingenio no tiene límites para transformar, optimizar, revolucionar…\n"
        "sin pausas y sin escalas. No espero la innovación… la construyo."
    )

    # Dividir el texto en líneas y dibujarlo
    for i, line in enumerate(paragraph.split("\n")):
        draw.text((text_x, text_y + i * 30), line, fill=(255, 255, 255), font=font)

    # Guardar la imagen final
    background.save(output_path)
    print(f"Banner guardado en {output_path}")

# Generar el banner
create_banner()
```
